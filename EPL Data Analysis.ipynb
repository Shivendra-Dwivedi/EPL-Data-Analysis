```python
import pandas as pd 
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('EPL_20_21.csv')
```

```python
df['MinsPerMatch'] = (df['Mins']/ df['Matches']).astype(int)
df['GoalsPerMatch'] = (df['Goals']/ df['Matches']).astype(float)
```

```python
df.head()
```

```python
Total_Goals = df['Goals'].sum()
print(Total_Goals)
```

```python
Penalty = df['Penalty_Goals'].sum()
print(Penalty)
```

```python
Penalty_Attempts = df['Penalty_Attempted'].sum()
print(Penalty_Attempts)
```

```python
plt.figure(figsize = (15,6))
pl_not_scored = df['Penalty_Attempted'].sum() - Penalty
data = [pl_not_scored, Penalty]
lables = ['Penalty Missed', 'Penalty Scored']
color = sns.color_palette('Set2')
plt.pie(data, labels = lables, colors = color, autopct = '%.0f%%')
plt.show()
```

```python
df['Club'].value_counts().nsmallest(10).plot(kind = 'bar', figsize = (12,6) , color = sns.color_palette('magma'))
```

```python
Under20 = df[df['Age'] <= 20]
age20_25 = df[(df['Age'] >20) &  (df['Age'] <=25)]
age25_30 = df[(df['Age'] >25) &  (df['Age'] <=30)]
Above_30 = df[df['Age'] >30]
```

```python
x = np.array([Under20['Name'].count(), age20_25['Name'].count(), age25_30['Name'].count(),Above_30['Name'].count()])
mylabels = ['Under20', '>20 & <=25','>25 & <=30', '>30']
plt.title('Players of Different Age Group', fontsize = 20)
plt.pie(x , labels = mylabels, autopct = '%.1f%%')
plt.show()
```

```python
plt.figure(figsize = (15,6))
sns.boxplot(x = 'Club', y =  'Age', data = df)
plt.xticks(rotation = 90)
```

```python
num_player = df.groupby('Club').size()
data = (df.groupby('Club')['Age'].sum())/num_player
data.sort_values(ascending = False)
```

```python
Assist = pd.DataFrame(df.groupby('Club', as_index = False)['Assists'].sum())
sns.set_theme(style = 'whitegrid', color_codes = True)
ax = sns.barplot(x = 'Club', y = 'Assists', data = Assist.sort_values(by = 'Assists'), palette = 'magma')
ax.set_xlabel('Clubs', fontsize = 30)
ax.set_ylabel('Assists', fontsize = 20)
plt.xticks(rotation = 90)
plt.rcParams['figure.figsize'] = (15,8)
plt.title('Plot of Clubs vs Total Assists', fontsize = 20)
```

```python
top_10_assist = df[['Name', 'Club', 'Assists', 'Matches']].nlargest(n = 10, columns = 'Assists')
top_10_assist
```

```python
plt.figure(figsize = (12,6))
assists = df['Assists'].sum()
data = (Total_Goals - assists, assists)
labels = ['Goals without Assist', 'Goals with Assist']
color = sns.color_palette('Set1')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```
